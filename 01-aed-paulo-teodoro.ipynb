```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregamento dos dados
train = pd.read_csv('train.csv')

# 2. Análise Básica e Verificação de Nulos
print("Informações do Dataset:")
print(train.info())
print("\nDistribuição das Classes:")
print(train['label'].value_counts())

# 3. Visualização da Distribuição das Classes
plt.figure(figsize=(8, 5))
sns.countplot(x='label', data=train, palette='viridis')
plt.title('Distribuição de Notícias Reais vs Fake')
plt.show()

# 4. Análise de tamanho de texto (Tratando dados nulos preventivamente)
train['text_len'] = train['text'].fillna('').apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.boxplot(x='label', y='text_len', data=train, palette='viridis')
plt.title('Distribuição do Número de Palavras por Classe')
plt.ylabel('Número de Palavras')
plt.show()
```
