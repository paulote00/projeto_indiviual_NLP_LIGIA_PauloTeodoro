```python
import pandas as pd
import joblib

# 1. Carregar Dados de Teste
test = pd.read_csv('test.csv')
test['total_content'] = test['title'].fillna('') + " " + test['text'].fillna('')

# 2. Carregar os Artefatos Serializados (Pesos)
print("Carregando o modelo treinado...")
model = joblib.load('modelo_fake_news.pkl')
vectorizer = joblib.load('vectorizer.pkl')

# 3. Transformação e Predição (Sem re-treinar o vetorizador!)
print("Vetorizando os dados de teste...")
X_test_tfidf = vectorizer.transform(test['total_content'])

print("Realizando as predições...")
predictions = model.predict(X_test_tfidf)

# 4. Geração do Arquivo de Submissão
# Certifique-se de que a coluna de ID no test.csv se chama 'id'
submission = pd.DataFrame({'id': test['id'], 'label': predictions})
submission.to_csv('submission_final.csv', index=False)
print("\nArquivo 'submission_final.csv' gerado com sucesso para o Kaggle!")
```
